```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
#heatmaps
data=np.random.random((13,16))
plt.imshow(data)
plt.title("2-D heat map")
plt.colorbar()
plt.show()
```

```python
data=np.random.random((13,16))
plt.imshow(data,cmap="magma")
plt.title("2-D heat map")
plt.colorbar()
plt.show()
```

```python
#heatamps using seaborn library
data=np.random.randint(low=14,
                       high=100,
                       size=(10,10))
hm=sns.heatmap(data=data,annot=True)
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
u = -1 - X**2 + Y
v = 1 + X - Y**2
plt.streamplot(X, Y, u, v)
plt.title('Streamline Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
