```python
x_pos=0
y_pos=0
x_direct=-1
y_direct=1
fig,ax=plt.subplots(figsize=(12,7))
ax.quiver(x_pos,y_pos,x_direct,y_direct)
ax.set_title('quiver plot with one arrow')
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
arrow1 = plt.arrow(0, 0, 1, 1, head_width=0.1, head_length=0.2, color='red')
arrow2 = plt.arrow(-0.5, 0.5, 0, -1, head_width=0.1, head_length=0.2, color='blue')
ax.add_patch(arrow1)
ax.add_patch(arrow2)
ax.set_title('Two Arrows')
plt.show()
```

```python
x_pos = [0, 0,1]
y_pos = [0, 0,1]
x_direct = [1, 0,-1]
y_direct = [1, -1,-1]
ax = plt.gca()
ax.quiver(x_pos, y_pos, x_direct, y_direct, scale=5)
ax.axis([-1.5, 1.5, -1.5, 1.5])
plt.show()
```
