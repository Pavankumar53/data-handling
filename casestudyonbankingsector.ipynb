```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'Age Group': ['<21', '21-24', '24-27', '27-30', '30-33','33-36','36-39','39-42','42-45','45-48','48-51','51-54','54-57','57-60','>60'],
    'Total Number of Loans': [9, 310, 511, 4000, 4568,5698,8209,8117,9000,7600,6000,4000,2000,788,6],
    'Number of Bad Loans': [2,14,20,172,169,188,197,211,216,152,84,64,26,9,0],
    'Number of Good Loans':[7,296,491,3828,4399,5510,802,7906,8784,7448,5916,3936,1974,779,6],
    '% of Bad Loans':['22.2%','4.5%','3.9%','4.3%','3.7%','3.3%','2.4%','2.0%','1.4%','1.6%','1.3%','1.1%','0.0%','5.8%','7.3%'],
    '% of Good Loans':['77.8%','95.5%','96.1%','95.7%','96.3%','95.8%','93.7%','92.9%','96.9%','99.5%','89.7%','45.9%','56.7%','98.4%','43.4%'],
}

df = pd.DataFrame(data)
print("DataFrame:")
print(df)

plt.figure(figsize=(8, 5))
plt.bar(df['Age Group'], df['Number of Good Loans'])
plt.xlabel('Age Group')
plt.ylabel('Number of bad Loans')
plt.title('% by Age Group')
plt.show()
```

```python
# Combine Number of Good Loans and Bad Loans into a single DataFrame
combined_df = pd.DataFrame({
    'Age Group': df['Age Group'],
    'Number of Good Loans': df['Number of Good Loans'],
    'Number of Bad Loans': df['Number of Bad Loans']
})

# Plotting
plt.figure(figsize=(10, 6))

# Plot for Good Loans
plt.bar(combined_df['Age Group'], combined_df['Number of Good Loans'], color='g', label='Good Loans')

# Plot for Bad Loans
plt.bar(combined_df['Age Group'], combined_df['Number of Bad Loans'], color='r', label='Bad Loans', bottom=combined_df['Number of Good Loans'])

plt.xlabel('Age Group')
plt.ylabel('Number of Loans')
plt.title('Number of Loans by Age Group')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
